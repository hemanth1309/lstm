```python
# Rename columns for convenience
fundamental_data.columns = ['Parameter'] + [f'Year_{year}' for year in range(2005, 2025)]
```

```python
fundamental_data
```

```python
fundamental_long = fundamental_data.melt(id_vars='Parameter', var_name='Year', value_name='Value')

# Extract the year from the column names
fundamental_long['Year'] = fundamental_long['Year'].str.extract(r'(\d{4})').astype(int)
fundamental_long['Month'] = 3
```

```python
fundamental_long
```

```python
# Pivot fundamental data
fundamental_pivot = fundamental_long.pivot(index=['Month'], columns='Parameter', values='Value').reset_index()

fundamental_pivot['Year'] = fundamental_pivot.apply(
    lambda row: pd.to_datetime(f"{row['Year']}-{row['Month']:02d}-01").strftime('%b-%Y'),
    axis=1
)
fundamental_pivot.drop(columns='Month', inplace=True)
fundamental_pivot.head()

data['Year'] = data['Date'].dt.year
data['Year'] = data['Date'].apply(
    lambda date: date.strftime('%b-%Y')
)
data.head()
```

```python
data.head()
```

```python
data['Month-Year'] = pd.to_datetime(data['Year'], format='%b-%Y')
fundamental_pivot['Month-Year'] = pd.to_datetime(fundamental_pivot['Year'], format='%b-%Y')

# Define a function to find the financial year range for a given stock date
def find_financial_year_range(stock_date, fundamental_df):
    # Calculate the start and end of the financial year range
    financial_year_start = pd.to_datetime(f'{stock_date.year}-03-01')
    financial_year_end = pd.to_datetime(f'{stock_date.year + 1}-03-01')

    # Check if the stock date is before March of the current year
    if stock_date < financial_year_start:
        financial_year_start = pd.to_datetime(f'{stock_date.year - 1}-03-01')
        financial_year_end = pd.to_datetime(f'{stock_date.year}-03-01')

    # Filter fundamental data for the relevant financial year range
    return fundamental_df[
        (fundamental_df['Month-Year'] >= financial_year_start) & (fundamental_df['Month-Year'] < financial_year_end)
    ]

# Initialize list to collect results
results = []

# Iterate over stock data and find matching fundamental data
for _, stock_row in data.iterrows():
    stock_date = stock_row['Month-Year']
    matching_fundamental = find_financial_year_range(stock_date, fundamental_pivot)
    for _, fundamental_row in matching_fundamental.iterrows():
        result = stock_row.to_dict()
        result.update(fundamental_row.drop('Month-Year').to_dict())
        results.append(result)

# Create a DataFrame from results
merged_data = pd.DataFrame(results)

# Save the merged data
merged_data.to_csv('wipro_test_data.csv', index=False)

# Print first few rows to verify
merged_data.head()
```

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('wipro_new_data.csv')
```

```python
features = data.drop(columns=['Date', 'Close','Year','Month-Year'])
targets = data[['Close']]
```

```python
features = features.apply(pd.to_numeric, errors='coerce')
targets = targets.apply(pd.to_numeric, errors='coerce')
```

```python
features = features.apply(pd.to_numeric, errors='coerce')
targets = targets.apply(pd.to_numeric, errors='coerce')
scaler_features = MinMaxScaler()
scaler_targets = MinMaxScaler()

features_scaled = scaler_features.fit_transform(features)
targets_scaled = scaler_targets.fit_transform(targets)
```

```python
def create_sequences(data, target, seq_length):
    xs, ys = [], []
    for i in range(len(data) - seq_length):
        x = data[i:i+seq_length]
        y = target[i+seq_length]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

seq_length = 60
X, y = create_sequences(features_scaled, targets_scaled, seq_length)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python

X_train = np.array(X_train)
X_train = np.expand_dims(features_scaled, axis=1)  # (samples, time_steps, features)
y_train = targets_scaled
```

```python
X_train = np.array(X_train)
X_train.shape
X_train
```

```python
# Checking for NaNs in X_train
nan_indices = np.where(np.isnan(X_train))
print("NaN indices:", nan_indices)

# Fill NaNs with the mean of the column or drop them
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train.reshape(-1, X_train.shape[2])).reshape(X_train.shape)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
import numpy as np

# Define the model
model = Sequential()

# Add LSTM layer with input shape
# input_shape=(1, 97) corresponds to (timesteps, features)
model.add(LSTM(100, input_shape=(1, 97), return_sequences=False))

# Dense layers for further learning
model.add(Dense(50, activation='relu'))
model.add(Dense(1))  # Output layer for predicting one value

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Fit the model
# Ensure y_train is reshaped to (-1, 1) if necessary
y_train_reshaped = np.reshape(y_train, (-1, 1))

model.fit(X_train, y_train_reshaped, epochs=20, batch_size=64, validation_split=0.2)

# Make predictions
predictions = model.predict(X_train)

# Print model summary
model.summary()
```
